```python
from sklearn import datasets
```

```python
cali=datasets.fetch_california_housing()
X=cali.data
Y=cali.target
```

```python
X.shape
```

```python
import pandas as pd
df=pd.DataFrame(X)
print(cali.feature_names)
df.columns=cali.feature_names
print(type(df.columns))
# add 1 extra feature/column of 2 dimensional
# print(type(df.MedInc))
# print(df.MedInc.count()) #count->20640
# df["Medinc_Medinc"]=df.MedInc**2
# X2=df.values
# print(X2)
# print(X2.shape) #X2.shape->(20640,9) as it is prev 8,
# add 1 feature becomes 9
df.describe()
# print(df.values[0])
# type(cali.feature_names)
# print(cali.feature_names[0])
```

```python
# to access a certain column in our dataframe
# first arg (:) signifies which rows we would like to index
# second arg("MedInc") lets us index the column
df.loc[:,"MedInc"]
# iloc gets integer val indices
df.iloc[:,0]
```

```python
for i in range(0,len(cali.feature_names)):
    val=df.iloc[:,i]**2
    ans=df.columns[i]+"_"+df.columns[i]
    df[ans]=val
df.describe()
```

```python
# access the value of first row first col->[0][0]th index value
print(df.values[0][0])
```

```python
from sklearn import model_selection
X2=df.values
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y,random_state=0)
X2_train, X2_test, Y2_train, Y2_test = model_selection.train_test_split(X2,Y,random_state=0)
print(X2_train.shape)
print(X2_test.shape)
print(Y2_train.shape)
print(Y2_test.shape)
print(type(X_train))
# X_train
```

```python
from sklearn.linear_model import LinearRegression
alg1=LinearRegression()
alg2=LinearRegression()
```

```python
alg1.fit(X_train,Y_train)
alg2.fit(X2_train,Y2_train)
```

```python
Y_pred=alg1.predict(X_test)
train_score=alg1.score(X_train,Y_train)
test_score=alg1.score(X_test,Y_test)
print("Train score: ",train_score)
print("Test score: ",test_score)
train2_score=alg2.score(X2_train,Y2_train)
test2_score=alg2.score(X2_test,Y2_test)
print("Train score: ",train2_score)
print("Test score: ",test2_score)
```
